```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds = pd.read_csv('housing.csv')
ds.head()
```

```python
ds.describe()
```

```python
ds.info()
```

```python
ds.shape
```

```python
# data visualization using seaborn

# Count plot
plt.figure(figsize=(25,5))
sns.countplot(ds['furnishingstatus'])
plt.show()
```

```python
sns.countplot(x="bedrooms", data=ds )
```

```python
sns.countplot(x="stories", data=ds )
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(ds['area'], ds['price'])
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Price vs. Area")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='bedrooms', data=ds)
plt.xlabel("Number of Bedrooms")
plt.ylabel("Count")
plt.title("Number of Bedrooms")
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(ds['furnishingstatus'].value_counts(), labels=ds['furnishingstatus'].value_counts().index, autopct='%1.1f%%')
plt.title("Furnishing Status")
plt.show()
```
