```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
dataSet = pd.read_csv('Students.csv')
dataSet.head()
```

```python
# nan values handling
m_mean = dataSet['MathScore'].mean()
e_mean = dataSet['EnglishScore'].mean()
dataSet['MathScore'].fillna(m_mean, inplace=True)
dataSet['EnglishScore'].fillna(m_mean, inplace=True)
dataSet
```

```python
# male and female , 0 and 1
le = LabelEncoder()
dataSet['Gender'] = le.fit_transform(dataSet['Gender'])
dataSet
```

```python
# correct score
dataSet['TotalScore'] = dataSet['MathScore'] + dataSet['EnglishScore']
dataSet
```

```python
scaler = StandardScaler()
dataSet[['MathScore', 'EnglishScore', 'TotalScore']] = scaler.fit_transform(dataSet[['MathScore', 'EnglishScore', 'TotalScore']])
dataSet
```
